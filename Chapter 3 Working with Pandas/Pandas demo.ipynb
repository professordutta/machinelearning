```python
# reload the data frame from the csv file
df = pd.read_csv('practice_dataset.csv')
df.info()
```

```python
# replace the missing values in calories column with the mean value of the calories column
df['Calories'].fillna(df['Calories'].mean(), inplace=True)
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'])
```

```python
# drop the row where there is missing value in the 'Date' column
df.dropna(subset=['Date'], inplace = True)
```
