```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
churn_df=pd.read_csv('churn_data.csv')
```

```python
churn_df.head()
```

```python
churn_df.describe()
```

```python
df=churn_df.drop('customerID',axis=1)
```

### As we can see TotalCharges column has object Dtype, let's change that into int

```python
df.TotalCharges.isnull().sum()
```

Firstly, let's drop all the rows containing empty string

```python
df = df[df.TotalCharges!=' ']
```

```python
df['TotalCharges']
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges)
```

```python
df['TotalCharges']
```

Now all the dtype of TotalCharges is float64

Let's create a function that helps us identifying object type columns

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df)
```

There are fields with values Yes, No and No internet service or No phone service. Let's quickly change No internet service or No phone service to No.

```python
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
df['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
print_unique_col_values(df)
```

Let's convert columns with Yes and No values to 1 and 0

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity',
                  'OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
                  'StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}') 
```

```python
#one hot encoding InternetService, Contract and PaymentMethod columns.

df_final = pd.get_dummies(data=df, columns=['InternetService','Contract','PaymentMethod'])
df_final.columns
```

```python
df_final.head()
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_final[cols_to_scale] = scaler.fit_transform(df_final[cols_to_scale])
```

```python
df_final.head()
```

```python
df_final.dtypes
```

```python
plt.figure(figsize=(15,14))
sns.heatmap(df_final.corr(),annot=True)
```

```python
X = df_final.drop('Churn',axis='columns')
y = df_final['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
