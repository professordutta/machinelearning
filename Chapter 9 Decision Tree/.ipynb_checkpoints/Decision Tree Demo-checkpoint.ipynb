# Perform EDA and Data Preprocessing

```python
# Find missing values using heat map
sns.heatmap(titanic.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# use count plot to check how many males and females surived
sns.countplot(x='Survived', hue='Sex', data=titanic)
```

```python
# use count plot to check how many people survived on the basis of Pclass
sns.countplot(x='Survived', hue='Pclass', data=titanic)
```

```python
# Do data preprocessing and work on missing values in the age column 
# use median age to fill the missing values
k = titanic['Age'].median()
print (k)
titanic['Age'] = titanic['Age'].fillna(k)
```

```python
embarked = pd.get_dummies(titanic['Embarked'], drop_first=False)
embarked
```

# Get the data that is needed for further analysis

```python
titanic.drop(['Embarked'], axis=1,  inplace=True)
titanic = pd.concat([titanic, embarked], axis=1)
titanic.head()
```

```python
# get the actual target values in y variable and variable X holding inputs
y = titanic[['Survived']].copy()
X = titanic[['Pclass','Sex','Age','Fare','C','Q','S']].copy()
print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
# Convert Sex column fields into binary form using label encoder
from sklearn.preprocessing import LabelEncoder
obj=LabelEncoder()
X['Sex']=obj.fit_transform(X['Sex'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
print(X_train.shape)
print(X_test.shape)
```

```python
X_train.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
