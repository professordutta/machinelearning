```python
df = pd.DataFrame(diabetes['data'],columns=diabetes['feature_names'])
```

```python
df.head()
```

```python
df.tail()
```

```python
# scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
diabetes_scaled = scaler.transform(df)
```

```python
# use PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
diabetes_scaled
```

```python
pca.fit(diabetes_scaled)
```

```python
xpca = pca.transform(diabetes_scaled)
```

```python
diabetes_scaled.shape
```

```python
xpca.shape
```

```python
xpca
```

```python
plt.scatter(xpca[:,0], xpca[:,1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
```

```python
pca.components_
```

```python
diabetes_new = pd.DataFrame(pca.components_, columns = diabetes['feature_names'])
```

```python
diabetes_new
```

```python
sns.heatmap(diabetes_new, cmap='plasma',annot=True)
```

# MODELLING AFTER PCA

```python
from sklearn.model_selection import train_test_split
```

```python
X = xpca
y = diabetes['target']
```

```python
xpca.shape
```

```python
diabetes['target'].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print(predictions[0:5]) 
```

```python
print(y_test[0:5])
```

```python
from sklearn import
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
```

```python
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
