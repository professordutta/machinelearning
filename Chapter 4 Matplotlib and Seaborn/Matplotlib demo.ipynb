## Basics 

```python
# lets create a numpy array for the values of x and y.
x = np.array([2,4,6,8,10])
y = np.array([3,6,9,12,15])
print(x,y)
print(type(x), type(y))
```

```python
# let plot the values of x and y with axis and title of the graph 
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Title of the graph")
```

## Subplots 

```python
# Subplots
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'b--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'y*-');
```

## Plot using the concept of OOPS

```python
# OOPS
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.2, 0.2, 0.7, 0.7]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('X axis') # Notice the use of set_ to begin methods
axes.set_ylabel('Y axis')
axes.set_title('Title of the graph');
```

## Styling 

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+2, color="black", alpha=0.5) # half-transparant
ax.plot(x, x+3, color="#8B008B")        # RGB hex code
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="purple", lw=3, linestyle='-')
ax.plot(x, x+6, color="purple", lw=3, ls='-.')
ax.plot(x, x+7, color="purple", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="green", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="green", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="green", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="red", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="red", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="red", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="yellow")
ax.plot(x, x+16, color="red", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="purple", markeredgewidth=3, markeredgecolor="green");
```

## Custom axis

```python
fig, ax = plt.subplots(figsize=(10, 4))
x = np.linspace(0, 5, 11)
y = x ** 2
ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

## Adding Labels

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="Label 1")
ax.plot(x, x**3, label="Label 2")
ax.legend()
```

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# Most common to choose where matplotlib decides the optimal location
ax.legend(loc=0) 
fig
```

## Adding text in the graph 

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**3, lw=1.5, color="green")
ax1.set_ylabel(r"Axis 1 ", fontsize=15, color="green")
ax1.text(2.5, 50.5, r"$y=x^3$", fontsize=18, color="green")
# ax1.text(left,height)
for label in ax1.get_yticklabels():
    label.set_color("green")
    
ax2 = ax1.twinx()
ax2.plot(x, x**5, lw=1.5, color="purple")
ax2.set_ylabel(r"Axis 2", fontsize=18, color="purple")
ax2.text(3.5, 12.5, r"$y=x^5$", fontsize=20, color="purple");
for label in ax2.get_yticklabels():
    label.set_color("purple")
```

```python
# axis spines, twin axis, axis where x and y is 0
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
```

```python
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
fig, ax = plt.subplots()

p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig, ax = plt.subplots()

cnt = ax.contour(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```
