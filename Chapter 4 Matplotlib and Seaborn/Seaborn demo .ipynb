```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the dataset
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

# Distribution Plots

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.rugplot(df['total_bill'])
```

```python
# displot - shows the distribution of univariate set of observations
```

```python
sns.distplot(df['total_bill'])
```

```python
# jointplot - used for combining two displots for bivariate data
# plots can be: scatter, reg, resid, kde, hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.pairplot(df,hue='sex')
```

```python
p = sns.pairplot(df)
p.map_diag(plt.hist)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(df, col="time",  row="smoker",hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

# Categorical Plots 

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

```python
sns.countplot(x='sex',data=df,hue='smoker')
sns.despine()
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='magma')
```

```python
sns.violinplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=df,hue='sex',palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=df,hue='sex',split=True,palette='rainbow')
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
# factor plot is a general plot that can take a "kind" parameter to adjust the plot type
```

```python
sns.factorplot(x='sex',y='total_bill',data=df,kind='bar')
```

# Matrix Plot

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
```

# Regression Plots

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex', markers=['o','v'],scatter_kws={'s':100})
```

# Aspect and size

```python
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=df)
```

# Using grids

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=df)
```
